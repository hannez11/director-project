# merge with director-portfolio-years

```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'

df1 = pd.read_csv(r"C:\Users\g21728\Desktop\Directors\Data\Hannes\director-year_marketcap_capiq_datastream_merge_120k.csv", decimal=",", sep=";")
df2 = pd.read_excel(r"C:\Users\g21728\Desktop\Directors\Data\NA - SMDEs Profile - Education & Achievements_cleaned.xlsx", sheet_name="Education")
df3 = pd.read_excel(r"C:\Users\g21728\Desktop\Directors\Data\NA - SMDEs Profile - Education & Achievements_cleaned.xlsx", sheet_name="Achievements")
```

```python
df1
```

```python
#assign ivyleague dummy if director attended at least one ivy league uni
ivy_league = "|".join(["Brown University", "Columbia University", "Cornell University", "Dartmouth College", "Harvard", "University of Pennsylvania", "Princeton", "Yale"])

df2["IvyLeague"] = df2["Institution Name"].str.contains(ivy_league).astype("Int64") #True if institution name contains any of the list items. convert boolean to 1 and 0
df2 = df2.drop(["Institution Name"], axis=1)

df2
```

```python
#groupby phd and ivyleague per director
df2 = df2.groupby(["DirectorID"]).max() #each director has multiple schools. reduce to one observation per director and set to one if at least one PhD received and one ivy league attended
df2
```

```python
#group by awards sum per director
df3 = df3.groupby(["DirectorID"]).sum()
df3
```

```python
#merge left (use first df as baseline. add columns from second df to baselind)
df_director_year_merge_education = df1.merge(df2, how="left", on="DirectorID")

df_director_year_merge_education_achievements = df_director_year_merge_education.merge(df3, how="left", on="DirectorID")

df_director_year_merge_education
df_director_year_merge_education_achievements
```

# +Characteristics

```python
df_c1 = pd.read_excel(r"C:\Users\g21728\Desktop\Directors\Data\Characteristics\clean\NA - SMDEs Profile - Characteristics - 1_clean.xlsx")
df_c2 = pd.read_excel(r"C:\Users\g21728\Desktop\Directors\Data\Characteristics\clean\NA - SMDEs Profile - Characteristics - 2_clean.xlsx")
df_c3 = pd.read_excel(r"C:\Users\g21728\Desktop\Directors\Data\Characteristics\clean\NA - SMDEs Profile - Characteristics - 3_clean.xlsx")
df_c4 = pd.read_excel(r"C:\Users\g21728\Desktop\Directors\Data\Characteristics\clean\NA - SMDEs Profile - Characteristics - 4_clean.xlsx")
```

```python
#combine to one df
df_characteristics = pd.concat([df_c1, df_c2, df_c3, df_c4])
df_characteristics
```

```python
#remove year and drop duplicate row observations
df_characteristics = df_characteristics.drop(labels="Year", axis=1).drop_duplicates()
df_characteristics['Gender'] = df_characteristics['Gender'].map({'M': 1, 'F': 0})
df_characteristics['DOB'].replace({"Unknown": 0}, inplace=True)
df_characteristics
```

```python
#convert all DOB entries to year ints
df_characteristics['DOB'] = df_characteristics['DOB'].astype(str).str[0:4] #take first 4 chars of date of birth
df_characteristics["DOB"] = df_characteristics["DOB"].astype("float").astype("Int64") #convert year to int
df_characteristics
```

```python
df_director_year_merge_education_achievements_characteristics = df_director_year_merge_education_achievements.merge(df_characteristics, how="left", on=["DirectorID"]).sort_values(by=["DirectorID", "Year"], ascending=[True,True], ignore_index=True)
#TypeError: unsupported operand type(s) for -: 'int' and 'datetime.datetime'
# df_director_year_merge_education_achievements_characteristics["Age"] = df_director_year_merge_education_achievements_characteristics["Year"] - df_director_year_merge_education_achievements_characteristics["DOB"]
# df_director_year_merge_education_achievements_characteristics
```
