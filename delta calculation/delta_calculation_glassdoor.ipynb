```python
from functools import reduce

#merge 3 gd frames
data_frames = [df_gd_agg, df_gd_sum, df_gd_count]
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=["DirectorID", "Year"], how='outer'), data_frames)

#merge with latest
df_merged = df_merged[df_merged["PF_GD_count_withEventFirm"] >= 2]
df_merged_gd = df_merged.merge(df_latest_firm_gd, how="left", on=["DirectorID", "Year"])

#cut column
col = df_merged_gd.pop("ISIN")
#insert column
df_merged_gd.insert(df_merged_gd.columns.get_loc("DirectorID") + 1, col.name, col)

#calculate (pf avg - firm rating) if dir-year obs has more than 2 firms in his gd pf
df_merged_gd["PF_GD_rating_overall_withoutEventFirm"] = np.where(df_merged_gd["PF_GD_count_withEventFirm"] >= 2, ( df_merged_gd["PF_GD_rating_overall_sum_withEventFirm"] - df_merged_gd["FIRM_GD_rating_overall"] ) / ( df_merged_gd["PF_GD_count_withEventFirm"] - 1 ), np.nan )

#calc delta firm minus pf rating
df_merged_gd["delta_FIRM_PF_GD_rating_overall"] = np.where(df_merged_gd["PF_GD_rating_overall_withoutEventFirm"].isna() == False, df_merged_gd["FIRM_GD_rating_overall"] - df_merged_gd["PF_GD_rating_overall_withoutEventFirm"], np.nan)

df_merged_gd[["FIRM_GD_rating_overall", "PF_GD_rating_overall_withoutEventFirm", "delta_FIRM_PF_GD_rating_overall"]]
```

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)

df_latest = pd.read_csv(r"D:\Docs\Dokumente\WMA\Projects\5 Directors\Data\Director-Firm-Years\0105director-firm-years_beta.csv", sep=";", decimal='.')
df_gainer = pd.read_excel(r"F:\Downloads\gain_happened_05012022.xlsx")
df_gainer.head()
```

```python
df_merge_gainer = df_latest.merge(df_gainer, how="left", on=["DirectorID", "FirmID", "ISIN", "Year"])
df_merge_gainer

#move new columns
gainer_cols = [col for col in df_gainer if "PF_" in col]
print(gainer_cols)
for index, column in enumerate(gainer_cols, start=1):
    #cut column
    col = df_merge_gainer.pop(column)
    #insert column
    df_merge_gainer.insert(df_merge_gainer.columns.get_loc("PF_gain_loss_replacement_indicator") + index, col.name, col)

df_merge_gainer
```

```python
df_pfcaptotal = pd.read_excel(r"F:\Downloads\RepTotal_05012022.xlsx")
df_pfcaptotal.head()
```

```python
df_merge = df_merge_gainer.merge(df_pfcaptotal, how="left", on=["DirectorID", "Year"])

df_merge.drop(columns=["PF_Ego_Market_Rep_Total"], inplace=True)

col = df_merge.pop("PF_MarketCap_Total_ego")
df_merge.insert(df_merge.columns.get_loc("PF_Ego_Market_Rep_Average") + 1, col.name, col)

df_merge.head()
```

```python
#calculate (pf avg - firm rating) if dir-year obs has more than 2 firms in his gd pf
df_merge["PF_MarketCap_withoutEvent_ego"] = np.where(df_merge["PF_no_of_dirs_selfcalc_ego"] >= 2, ( df_merge["PF_MarketCap_Total_ego"] - df_merge["FIRM_market_cap_total"] ) / ( df_merge["PF_no_of_dirs_selfcalc_ego"] - 1 ), np.nan )

#if event firm is not present in PF during a specific year, then just calc average based on existing PF / number of dirs (without adjusting by minus 1)
df_merge["PF_MarketCap_withoutEvent_ego"] = np.where(df_merge["FIRM_presence_in_PF"].isna() == True, ( df_merge["PF_MarketCap_Total_ego"] ) / ( 1 ), df_merge["PF_MarketCap_withoutEvent_ego"] )

#calc delta firm minus pf rating
df_merge["delta_MarketCap_ego"] = np.where(df_merge["PF_MarketCap_withoutEvent_ego"].isna() == False, df_merge["FIRM_market_cap_total"] - df_merge["PF_MarketCap_withoutEvent_ego"], np.nan)

#calc dummy: 1 if delta > 0 ; 0 if delta < 0
df_merge.insert(df_merge.columns.get_loc("delta_MarketCap_ego") + 1, "delta_MarketCap_ego_dummy", np.nan)
df_merge["delta_MarketCap_ego_dummy"] = np.where((df_merge["delta_MarketCap_ego"].isna() == False) & (df_merge["delta_MarketCap_ego"] > 0), 1, 
    np.where((df_merge["delta_MarketCap_ego"].isna() == False) & (df_merge["delta_MarketCap_ego"] < 0), 0, np.nan))

for index, column in enumerate(["PF_MarketCap_withoutEvent_ego", "delta_MarketCap_ego", "delta_MarketCap_ego_dummy"], start=1):
    #cut column
    col = df_merge.pop(column)
    #insert column
    df_merge.insert(df_merge.columns.get_loc("PF_MarketCap_Total_ego") + index, col.name, col)

df_test = df_merge[["DirectorID", "ISIN", "Year", "FIRM_presence_in_PF", "PF_no_of_dirs_selfcalc_ego", "FIRM_market_cap_total", "PF_MarketCap_Total_ego", "PF_MarketCap_withoutEvent_ego", "delta_MarketCap_ego", "delta_MarketCap_ego_dummy"]]
df_test[(df_test["PF_no_of_dirs_selfcalc_ego"] >= 1) & (df_test["delta_MarketCap_ego"] == 0)] #weird cases with delta = 0 that can be ignored
```
