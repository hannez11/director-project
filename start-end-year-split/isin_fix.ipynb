# correct long firmIDs with wrong 0s (caused by csv decimal . vs , issues)

```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'
pd.set_option('display.max_columns', None)

df_latest = pd.read_csv(r"D:\Docs\Dokumente\WMA\Projects\5 Directors\Data\Director-Firm-Years\1230director-firm-years_with-is_member_wrong_12digit_ids.csv", sep=";", decimal=".")
df_correct_firmid = pd.read_csv(r"D:\Docs\Dokumente\WMA\Projects\5 Directors\Data\Director-Firm-Years\1215director-firm-years-years_before_gain_wrong_decimal.csv", sep=";", decimal=",")

df_correct_firmid.rename(columns={"ISIN First": "ISIN"}, inplace=True)
print(df_correct_firmid.iloc[1145,1]) #check if complete 12 digit number
df_correct_firmid.drop_duplicates(inplace=True)

df_correct_firmid
```

### correct firmid isin pairs

```python
df_correct_firmid_without_dirid = df_correct_firmid[["FirmID", "ISIN"]]
df_correct_firmid_without_dirid.drop_duplicates(inplace=True, keep="first")
df_correct_firmid_without_dirid
```

```python
#check if any duplicate dir firm year entries exist
df_latest["check"] = df_latest.duplicated(keep=False) #False : Mark all duplicates as True

#display duplicates. should be empty
df_latest[df_latest["check"] == True]

#remove duplicates
df_latest.drop(columns=["check"], inplace=True)
df_latest.drop_duplicates(keep="first", inplace=True)
df_latest
```

```python
#merge

df_latest_correct_firmid = df_latest.merge(df_correct_firmid_without_dirid, how="left", on=["ISIN"], suffixes=("_wrong", "_correct"))

#move column
col = df_latest_correct_firmid.pop("FirmID_correct")
#insert column
df_latest_correct_firmid.insert(df_latest_correct_firmid.columns.get_loc("FirmID_wrong")+1, col.name, col)

df_latest_correct_firmid
```

```python
#check if any duplicate dir firm year entries exist
df_latest_correct_firmid["check"] = df_latest_correct_firmid.duplicated(subset=["DirectorID", "ISIN", "Year"], keep=False) #False : Mark all duplicates as True
#display duplicates. should be empty
df_duplicate_rows = df_latest_correct_firmid[df_latest_correct_firmid["check"] == True].sort_values(by=["DirectorID", "ISIN", "Year"], ascending=[True,True, True], ignore_index=True)
# df_duplicate_rows
# df_duplicate_rows.to_csv(r"C:\Users\hannes\Desktop\df_duplicate_rows_isin.csv", sep=";", decimal='.', index=False)

#remove duplicates
df_latest_correct_firmid.drop(columns=["check"], inplace=True)
df_latest_correct_firmid.drop_duplicates(subset=["DirectorID", "ISIN", "Year"], keep="last", inplace=True)
df_latest_correct_firmid
```

### correct directorids

```python
df_correct_firmid_with_dirid = df_correct_firmid[["DirectorID", "FirmID"]]
df_correct_firmid_with_dirid.drop_duplicates(inplace=True)

#filter dirIDs with len >= 12 (e+11)
df_correct_firmid_with_dirid["DirectorID"] = df_correct_firmid_with_dirid["DirectorID"].astype("str")

#insert new column with first 3 chars of directorid string
if 'DirectorID_string' not in df_correct_firmid_with_dirid.columns:
    df_correct_firmid_with_dirid.insert(df_correct_firmid_with_dirid.columns.get_loc("DirectorID")+1, "DirectorID_string", np.nan)
df_correct_firmid_with_dirid["DirectorID_string"] = df_correct_firmid_with_dirid["DirectorID"].astype("str").str[:4]

df_correct_firmid_with_dirid_firstthree = df_correct_firmid_with_dirid[df_correct_firmid_with_dirid["DirectorID"].apply(lambda x: len(x)>=12)]
df_correct_firmid_with_dirid_firstthree.rename(columns={"FirmID": "FirmID_correct"}, inplace=True)

#int with 3 first digits is rounded based on fourth digit
if 'DirectorID_fourth_digit' not in df_correct_firmid_with_dirid_firstthree.columns:
    df_correct_firmid_with_dirid_firstthree.insert(df_correct_firmid_with_dirid_firstthree.columns.get_loc("DirectorID_string")+1, "DirectorID_fourth_digit", np.nan)
    df_correct_firmid_with_dirid_firstthree.insert(df_correct_firmid_with_dirid_firstthree.columns.get_loc("DirectorID_string")+2, "DirectorID_rounded", np.nan)

#get fourth digit 
df_correct_firmid_with_dirid_firstthree["DirectorID_fourth_digit"] = df_correct_firmid_with_dirid_firstthree["DirectorID_string"].str[-1:]
df_correct_firmid_with_dirid_firstthree["DirectorID_string"] = df_correct_firmid_with_dirid_firstthree["DirectorID"].astype("str").str[:3]

df_correct_firmid_with_dirid_firstthree
```

```python
#transform back to ints
for column in ["DirectorID_string", "DirectorID_fourth_digit"]:
    df_correct_firmid_with_dirid_firstthree[column] = df_correct_firmid_with_dirid_firstthree[column].astype("int64")

#add 1 if fourth digit >= 5
df_correct_firmid_with_dirid_firstthree["DirectorID_rounded"] = np.where(df_correct_firmid_with_dirid_firstthree["DirectorID_fourth_digit"] >= 5, df_correct_firmid_with_dirid_firstthree["DirectorID_string"] + 1, df_correct_firmid_with_dirid_firstthree["DirectorID_string"])

df_correct_firmid_with_dirid_firstthree.drop(columns=["DirectorID_string", "DirectorID_fourth_digit"], inplace=True)
df_correct_firmid_with_dirid_firstthree.rename(columns={"DirectorID_rounded": "DirectorID_string"}, inplace=True)
df_correct_firmid_with_dirid_firstthree["DirectorID_string"] = df_correct_firmid_with_dirid_firstthree["DirectorID_string"].astype("str")

df_correct_firmid_with_dirid_firstthree
```

```python
#insert column if not already existing
if 'DirectorID_string' not in df_latest_correct_firmid.columns:
    df_latest_correct_firmid.insert(df_latest_correct_firmid.columns.get_loc("DirectorID")+1, "DirectorID_string", np.nan)

df_latest_correct_firmid["DirectorID_string"] = df_latest_correct_firmid["DirectorID"].astype("str").str[:3]

df_latest_wrong_directorid = df_latest_correct_firmid[df_latest_correct_firmid["DirectorID"].astype("str").apply(lambda x: len(x)>=14)]
df_latest_wrong_directorid
```

```python
df_latest_correct_firmid_dirid = df_latest_correct_firmid.merge(df_correct_firmid_with_dirid_firstthree, how="left", on=["FirmID_correct", "DirectorID_string"], suffixes=("_wrong", "_correct"))

#move column
col = df_latest_correct_firmid_dirid.pop("DirectorID_correct")
#insert column
df_latest_correct_firmid_dirid.insert(df_latest_correct_firmid_dirid.columns.get_loc("DirectorID_wrong")+1, col.name, col)
df_latest_correct_firmid_dirid.drop(columns=["DirectorID_string"])

#sort
df_latest_correct_firmid_dirid = df_latest_correct_firmid_dirid.sort_values(by=["DirectorID_wrong", "FirmID_correct", "Year"], ascending=[True,True, True], ignore_index=True)

#combine correct existing director values and new values in the new column
df_latest_correct_firmid_dirid["DirectorID_correct"] = np.where(df_latest_correct_firmid_dirid["DirectorID_correct"].isna(), 
    df_latest_correct_firmid_dirid["DirectorID_wrong"], 
    df_latest_correct_firmid_dirid["DirectorID_correct"])

df_latest_correct_firmid_dirid["DirectorID_correct"] = df_latest_correct_firmid_dirid["DirectorID_correct"].astype("int64")

df_latest_correct_firmid_dirid.drop(columns=["DirectorID_wrong", "DirectorID_string", "FirmID_wrong"], inplace=True)
df_latest_correct_firmid_dirid.rename(columns={"DirectorID_correct": "DirectorID", "FirmID_correct": "FirmID"}, inplace=True)

df_latest_correct_firmid_dirid
```

### check for duplicates

```python
#check if any duplicate dir firm year entries exist
df_latest_correct_firmid_dirid["check"] = df_latest_correct_firmid_dirid.duplicated(subset=["DirectorID", "ISIN", "Year"], keep=False) #False : Mark all duplicates as True
#display duplicates. should be empty
df_duplicate_rows = df_latest_correct_firmid_dirid[df_latest_correct_firmid_dirid["check"] == True].sort_values(by=["DirectorID", "ISIN", "Year"], ascending=[True,True, True], ignore_index=True)
df_duplicate_rows

#remove duplicates
df_latest_correct_firmid_dirid.drop(columns=["check"], inplace=True)
df_latest_correct_firmid_dirid.drop_duplicates(subset=["DirectorID", "ISIN", "Year"], keep=False, inplace=True) #weird duplicates and only 40 obs -> drop them all
df_latest_correct_firmid_dirid
```
