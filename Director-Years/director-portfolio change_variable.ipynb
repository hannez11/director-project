# Director-Portfolio-Years
how did the yearly director portfolio composition change?

```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'

df = pd.read_excel(r"C:\Users\g21728\Desktop\Directors\Data\Director-year-ISIN_Panel.xlsx")
df
```

```python
directorships = df.iloc[:, [i for i in range(2,len(df.columns))]].values.tolist() # convert rows to lists. start at third column

directorships_cleaned = list() # remove nan and sort alphabetically
for i in directorships:
    new_i = list()
    new_i = [str(x) for x in i if type(x) is not float] # remove nan (stored as floats)
    directorships_cleaned.append(sorted(new_i))

# store NaNs at the end of the list
df2 = df.iloc[:, [0,1]] # empty dataframe with just headers

helper_cols = ["PF_changed", "previous", "new", "change"]
for i in helper_cols:
    df2[i] = np.nan 
    df2[i] = df2[i].astype('object') # column has to be object type, otherwise lists cannot be stored later on in their cells

# add X empty columns to dataframe
max_simultaneous_directorships = df.count(axis=1).max() - 2 # get highest value of non nan row values (minus 2 to ignore first two columns)
# print(max_simultaneous_directorships)
for i in range(1,max_simultaneous_directorships+1): 
    df2[i] = np.nan

for row, i in enumerate(directorships_cleaned): # place directorship lists in empty columns
    df2.loc[row, [j for j in range(1, len(i)+1)]] = i # start at first empty column
    
#probably faster with apply:
# split list items of previous column in X new columns (better approach)
# df_previous = df3.previous.apply(pd.Series).add_prefix('previous_')

df2
```

```python
dirs_cleaned = list() # empty list to store directorships per director-year, without NaNs
dirs = df2.iloc[:, [i for i in range(0,len(df2.columns))]].values.tolist() # convert rows to lists. start at third column
for i in dirs:
    # print([x for x in i if type(x) is not float])
    new_i = list()
    new_i = [str(x) for x in i if type(x) is not float]
    dirs_cleaned.append(new_i)

# print(dirs_cleaned)

for i, v in enumerate(dirs_cleaned, start=0): #skip first row since value is compared with previous value

    previous_portfolio = dirs_cleaned[i-1][2:]
    new_portfolio = dirs_cleaned[i][2:]
    
    if new_portfolio == previous_portfolio: # if portfolio didnt change
        df2.at[i, "PF_changed"] = 0

    elif (dirs_cleaned[i][0] == dirs_cleaned[i-1][0]) and new_portfolio != previous_portfolio: # same director and change in portfolio
        uniques = list(set(new_portfolio) ^ set(previous_portfolio)) # XOR operator to get difference between two sets. transform set to list
        if len(new_portfolio) > len(previous_portfolio):
            if len(new_portfolio) - len(previous_portfolio) != len(uniques):
                df2.iloc[i, 2] = "gain doublecheck" # set binary to 1 that pf increased
            else:
                df2.iloc[i, 2] = "gain" # set binary to 1 that pf increased
        elif len(new_portfolio) < len(previous_portfolio):
            if len(new_portfolio) < len(previous_portfolio):
                if len(previous_portfolio) - len(new_portfolio) != len(uniques):
                    df2.iloc[i, 2] = "loss doublecheck" # set binary to 1 that pf increased
                else:
                    df2.iloc[i, 2] = "loss" # set binary to 2 that pf decreased
        else:
            df2.iloc[i, 2] = "replace" # change occurred, without changing length of the two portfolios (e.g., one dir got replaced by another)

        # get change values for change column
        # print("uniques", uniques, "\n")
        if len(uniques) > 0: # if uniques not empty
            for ind,val in enumerate(uniques): # i,v already used
                if val in previous_portfolio:
                    uniques[ind] = val + "-" # if unique value was in previous PF, then dir was lost -> add a minus
                else:
                    uniques[ind] = val + "+" # if unique value is in new PF, then dir was gained -> add a plus
        df2.at[i, "change"] = uniques

    # get previous portfolio values for previous column
    if (dirs_cleaned[i][0] == dirs_cleaned[i-1][0]):
        df2.at[i, "previous"] = previous_portfolio

    # get next portfolio values for next column
    df2.at[i, "new"] = new_portfolio

#check weird cases
# df2[df2["PF_changed"] == "gain doublecheck" | df2["PF_changed"] == "loss doublecheck"]

df2
```

```python
# split previous & change PF ISINs in list to new columns

df3 = df2.iloc[:, range(0,6)] # remove ISIN columns

# df3["previous"].map(lambda x: print(x), na_action=None) # display items of previous columns

# split list items of previous column in X new columns (better approach)
df_previous = df3.new.apply(pd.Series).add_prefix('new_')
df_change = df3.change.apply(pd.Series).add_prefix('change_')
df_previous_change = df_previous.merge(df_change, how="outer", left_index=True, right_index=True) # merge previous and change df

df4 = df3.merge(df_previous_change, how="outer", left_index=True, right_index=True) # merge with prior df

df4
```
