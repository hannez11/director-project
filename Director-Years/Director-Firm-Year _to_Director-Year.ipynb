# Director-Years

#### directorships [15 Jan]

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)

#load
df_latest = pd.read_csv(r"D:\Docs\Dokumente\WMA\Projects\5 Directors\Data\Director-Firm-Years\0115cdirector-firm-years_iota.csv", sep=";", decimal='.')
#reorder columns
sorted_columns = df_latest.columns.to_list()[:20] + sorted(df_latest.columns.to_list()[20:], key=str.casefold) #keep first 20 columns. sort all others alphabetically & ignore case
df_latest = df_latest.reindex(columns=sorted_columns)

#calc correct directorships
df_dirs = df_latest.groupby(["DirectorID", "Year"], as_index=False)["DIR_FIRM_active"].sum().rename(columns={"DIR_FIRM_active": "DIR_directorships_new"})
df_dirs["DIR_directorships_new"].replace({0: np.nan}, inplace=True)

df_latest = df_latest.merge(df_dirs, how="left", on=["DirectorID", "Year"])

#move columns
col = df_latest.pop("DIR_directorships_new")
df_latest.insert(df_latest.columns.get_loc("DIR_directorships") + 1, col.name, col)

#calc change in directorships
df_latest.insert(df_latest.columns.get_loc("DIR_directorships_new") + 1, "DIR_directorships_new_change", np.nan)
cond_dir_firmid_same = ( (df_latest["DirectorID"].shift(1) == df_latest["DirectorID"]) & (df_latest["FirmID"].shift(1) == df_latest["FirmID"]) )
df_latest["DIR_directorships_new_change"] = np.where(cond_dir_firmid_same & (df_latest["DIR_directorships_new"].isna() == False) & df_latest["DIR_directorships_new"].shift(1).isna() == False, 
    df_latest["DIR_directorships_new"] - df_latest["DIR_directorships_new"].shift(1), np.nan)

df_latest
```

```python
#drop firm columns
firm_columns = ["FirmID", "ISIN"] + [column for column in df_latest.columns.tolist() if "FIRM" in column]
df_reduced = df_latest.drop(columns=firm_columns).drop_duplicates(subset=["DirectorID", "Year"], keep="first", ignore_index=True) #only keep first dir-year observation

#create copy
df = df_reduced.copy()
df_reduced
```

## change variables (t / t-1)

```python
#calc avg based on sum / directorships for board_comp beta meetings
df["DIR_board_comp_avg"] = np.where((df["DIR_board_comp_sum"].isna() == False), df["DIR_board_comp_sum"] / df["DIR_directorships"], np.nan) #directorships might be outdated
df["DIR_board_comp_avg"].replace({0: np.nan}, inplace=True) #set 0 to np.nan, otherwise we divide with 0 sometimes

df["DIR_beta_avg_new"] = np.where((df["DIR_beta_sum"].isna() == False), df["DIR_beta_sum"] / df["DIR_directorships"], np.nan) #directorships might be outdated
df["DIR_beta_avg_new"].replace({0: np.nan}, inplace=True) #set 0 to np.nan, otherwise we divide with 0 sometimes

df["DIR_meetings_avg_new"] = np.where((df["DIR_meetings_sum"].isna() == False), df["DIR_meetings_sum"] / df["DIR_directorships"], np.nan) #directorships might be outdated
df["DIR_meetings_avg_new"].replace({0: np.nan}, inplace=True) #set 0 to np.nan, otherwise we divide with 0 sometimes

#calc relative change by t / t-1
df["DIR_board_comp_avg_change_rel"] =  np.where( (df["DirectorID"].shift(1) == df["DirectorID"]) & (df["DIR_board_comp_avg"].isna() == False) & (df["DIR_board_comp_avg"].shift(1).isna() == False), df["DIR_board_comp_avg"]/df["DIR_board_comp_avg"].shift(1), np.nan)

df["DIR_beta_avg_new_change_rel"] =  np.where((df["DirectorID"].shift(1) == df["DirectorID"]) & (df["DIR_beta_avg_new"].isna() == False) & (df["DIR_beta_avg_new"].shift(1).isna() == False), df["DIR_beta_avg_new"]/df["DIR_beta_avg_new"].shift(1), np.nan)

df["DIR_meetings_avg_new_change_rel"] =  np.where((df["DirectorID"].shift(1) == df["DirectorID"]) & (df["DIR_meetings_avg_new"].isna() == False) & (df["DIR_meetings_avg_new"].shift(1).isna() == False), df["DIR_meetings_avg_new"]/df["DIR_meetings_avg_new"].shift(1), np.nan)

#calc absolute change by t - t-1
df["DIR_board_comp_avg_change_abs"] =  np.where( (df["DirectorID"].shift(1) == df["DirectorID"]) & (df["DIR_board_comp_avg"].isna() == False) & (df["DIR_board_comp_avg"].shift(1).isna() == False), df["DIR_board_comp_avg"]-df["DIR_board_comp_avg"].shift(1), np.nan)

df["DIR_beta_avg_new_change_abs"] =  np.where((df["DirectorID"].shift(1) == df["DirectorID"]) & (df["DIR_beta_avg_new"].isna() == False) & (df["DIR_beta_avg_new"].shift(1).isna() == False), df["DIR_beta_avg_new"]-df["DIR_beta_avg_new"].shift(1), np.nan)

df["DIR_meetings_avg_new_change_abs"] =  np.where((df["DirectorID"].shift(1) == df["DirectorID"]) & (df["DIR_meetings_avg_new"].isna() == False) & (df["DIR_meetings_avg_new"].shift(1).isna() == False), df["DIR_meetings_avg_new"]-df["DIR_meetings_avg_new"].shift(1), np.nan)

#subsample with relative changes
df[df["DIR_beta_avg_new"].isna() == False][["DirectorID", "DIR_board_comp_avg", "DIR_board_comp_avg_change_rel", "DIR_beta_avg", "DIR_beta_avg_new", "DIR_beta_avg_new_change_rel", "DIR_meetings_avg_new", "DIR_meetings_avg_new_change_rel"]]
```

```python
#rename mcap columns
marketcap_cols = [col for col in df.columns.tolist() if "MarketCap" in col]
df.rename(columns={'DIR_MarketCap_avg': 'DIR_mcap_avg',
    'DIR_MarketCap_avg_change': 'DIR_mcap_avg_change_abs',
    'DIR_MarketCap_avg_change_rel': 'DIR_mcap_avg_change_rel',
    'DIR_MarketCap_avg_change_rel_dmy': 'DIR_mcap_avg_change_rel_dmy',
    'DIR_MarketCap_total': 'DIR_mcap_total',
    'DIR_MarketCap_total_withoutEvent': 'DIR_mcap_total_withoutEvent'}, inplace=True)

df
```

```python
#reorder columns
sorted_columns = df.columns.to_list()[:20] + sorted(df.columns.to_list()[20:], key=str.casefold) #keep first 20 columns. sort all others alphabetically & ignore case
df = df.reindex(columns=sorted_columns)

#export
df.to_csv(r"D:\Docs\Dokumente\WMA\Projects\5 Directors\Data\Director-Years\0115c_Director-Years.csv", sep=";", decimal='.', index=False)
```
